```python
import pandas as pd
import hashlib as hl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file = pd.read_csv("../athlete_events.csv")

# anonymisera kolumnerna med idrotternas namn
df = file.copy()
df["Name"] = df["Name"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

print(f" Kolumner som anonymiseras: {df.columns}")
```

```python
#  ldersfrdelning i sporterna
df_sporter = df[df["Sport"].isin(["Cross Country Skiing", "Football", "Sailing", "Handball"])]
df_sporter = df_sporter.dropna()
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=2, kde=True)

plt.title("ldersfrdelning i sporterna i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")
plt.show()
```

```python
#  ldersfrdelning i OS Lngdskidor
plt.figure(figsize=(12,12))
plt.subplot(3, 2, 1)
df_sporter = df[df["Sport"].isin(["Cross Country Skiing"])]
df_sporter = df_sporter.dropna()
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True, palette=["Orange"])

plt.title("ldersfrdelning i Lngdskidor i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")

plt.subplot(3, 2, 2)
df_sporter = df[df["Sport"].isin(["Football"])]
df_sporter = df_sporter.dropna()
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True)

plt.title("ldersfrdelning i Fotboll i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")

plt.subplot(3, 2, 3)
df_sporter = df[df["Sport"].isin(["Sailing"])]
df_sporter = df_sporter.dropna()
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True, palette=["Green"])

plt.title("ldersfrdelning i Segling i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare i procent")

plt.subplot(3, 2, 4)
df_sporter = df[df["Sport"].isin(["Handball"])]
df_sporter = df_sporter.dropna()
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True, palette=["Red"])

plt.title("ldersfrdelning i Handboll i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")

plt.tight_layout()
plt.show()
```

```python
df_sporter = df[df["Sport"].isin(["Cross Country Skiing", "Football", "Sailing", "Handball"])]
df_sporter = df_sporter.dropna()

# Skapa en figur
fig, ax = plt.subplots()

# Skapa histogrammet
bins = np.arange(df_sporter["Age"].min(), df_sporter["Age"].max() + 2, 2)
ax.hist(df_sporter["Age"], bins=bins, alpha=0.5, label="Cross Country Skiing")
ax.hist(df_sporter[df_sporter["Sport"] == "Football"]["Age"], bins=bins, alpha=0.5, label="Football")
ax.hist(df_sporter[df_sporter["Sport"] == "Sailing"]["Age"], bins=bins, alpha=0.5, label="Sailing")
ax.hist(df_sporter[df_sporter["Sport"] == "Handball"]["Age"], bins=bins, alpha=0.5, label="Handball")

# Lgg till titel, x- och y-etiketter
ax.set_title("ldersfrdelning i sporterna i OS")
ax.set_xlabel("lder")
ax.set_ylabel("Antal deltagare")

# Lgg till en legend
ax.legend()

# Visa figur
plt.show()
```
