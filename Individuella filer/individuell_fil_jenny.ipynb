```python
import pandas as pd
import hashlib as hl
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../athlete_events.csv")

# anonymisera kolumnerna med idrotternas namn
df_anonym = df.copy()
df_anonym["Name"] = df_anonym["Name"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

print(f" Kolumner som anonymiseras: {df_anonym.columns}")
```

```python
#  ldersfrdelning i sporterna
df_sporter = df_anonym[df_anonym["Sport"].isin(["Cross Country Skiing", "Football", "Sailing", "Handball"])]
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=2, kde=True)

plt.title("ldersfrdelning i sporterna i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")
plt.show()
```

```python
#  ldersfrdelning i OS Lngdskidor
df_sporter = df_anonym[df_anonym["Sport"].isin(["Cross Country Skiing"])]
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True, palette=["Orange"])

plt.title("ldersfrdelning i Lngdskidor i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")
plt.show()
```

```python
#  ldersfrdelning i OS Fotboll
df_sporter = df_anonym[df_anonym["Sport"].isin(["Football"])]
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True)

plt.title("ldersfrdelning i Fotboll i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare")
plt.show()
```

```python
#  ldersfrdelning i OS Segling
df_sporter = df_anonym[df_anonym["Sport"].isin(["Sailing"])]
sns.set_theme(style="darkgrid")
sns.histplot(x="Age", hue="Sport", data=df_sporter, binwidth=1, kde=True, palette=["Green"])

plt.title("ldersfrdelning i Segling i OS")
plt.xlabel("lder")
plt.ylabel("Antal deltagare i procent")

plt.show()
```
